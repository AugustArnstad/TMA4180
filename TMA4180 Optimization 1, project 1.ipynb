## 2.2 c)

We recall the definition of a convex function:

A function $f:S\rightarrow\mathbb{R}$ is convex if its domain $S\subseteq \mathbb{R^n}$ is a convex set and if for any two points $x, y \in S$ and $\lambda \in [0, 1]$ the following inequality holds:
$$f(\lambda x+(1-\lambda)y) \leq  \lambda f(x) + (1-\lambda)f(y)$$

We assume it to be known that $\mathbb{R^n}$ is a convex set. We further proceed by checking the inequality. 
Let V be a vectorfield, then a norm is a function $f:V\rightarrow\mathbb{R}$.

It is known that every norm satisfies the triangle inequality 
$$\lVert x+y\rVert\leq \lVert x \rVert + \lVert y \rVert$$
and the positive homogenity criteria, so we apply this to the inequality above:

$$\lVert \lambda x + (1-\lambda)y\rVert \leq |\lambda| \lVert x \rVert + |(1-\lambda)| \lVert y \rVert$$
which satisfies the inequality from the definition of a convex function. Hence every norm is a convex function. $\blacksquare$

## 2.2 4)

As we have established, every norm is convex. 
Further, if a norm $\lVert \cdot \rVert : \mathbb R^2 \rightarrow \mathbb R$ is a norm, then we can define a metric on $\mathbb R^2$ that is induced by the norm $\lVert \cdot \rVert$ as 
$$
d(x, y):=\lVert x-y \rVert
$$
We shall take advantage of this, by proving that the objective functions can be represented as norms, making the convexity trivial. 

#### Objective function 2.1):
We make the assumption that $A\subseteq \mathbb R^2$. Further we have 
$$
f(x)=\max_{a\in A} d(a,x)
$$
Recall the definition of p-norms on an n-dimensional (in our case $n=2$) vector space:
$$
\lVert x \rVert_p = \left(\sum_{i=1}^{n} x_i^{\frac{1}{p}}\right)^{p}=\left(\sum_{i=1}^{2} x_i^{\frac{1}{p}}\right)^{p}
$$
for $1\leq p < \infty$, and 
$$
\lVert x \rVert_{\infty}:=\max_{i} |x_i|
$$

We now notice that 
$$
f(x)=\max_{a\in A} d(a,x)= \max_{a\in A} \lVert a-x \rVert
$$
Hence $f(x)$ must be convex, as the the convexity holds for all points $a\in A$ and $x\in \mathbb R^2$, including the points with the maximal distance, which by definition is our $f(x)$.

#### Objective function 2.2):
With the same assumptions, we rewrite the objective function
$$
f(x)=\sum_{a\in A} d(a,x)=\sum_{a\in A} \lVert a-x \rVert
$$
and we procede to prove that the sum of norms satisfy the properties of a norm;
$$
\lVert x \rVert_a + \lVert y \rVert_b \geq 0 \\
|\alpha|\left(\lVert x \rVert_a + \lVert y \rVert_b\right) = |\alpha|\lVert x \rVert_a + |\alpha|\lVert y \rVert_b=\lVert \alpha x \rVert_a + \lVert \alpha y \rVert_b  \\
\lVert x + z \rVert_a + \lVert y + z \rVert_b \leq \lVert x \rVert_a + \lVert z \rVert_a + \lVert y \rVert_b + \lVert z \rVert_b
$$
where the triangle inequality holds immediately as both $\lVert \cdot \rVert_a, \lVert \cdot \rVert_b$ are norms by assumption.

We conclude that also this objective function can be written as a norm, and so it must be convex. $\blacksquare$

We begin by recalling that the Euclidean norm is convex. Going forward, we can find from the definition of a convex function
$$
f(\lambda x+(1-\lambda)y) \leq  \lambda f(x) + (1-\lambda)f(y) \\
f^2(\lambda x+(1-\lambda)y) \leq  \lambda f^2(x) + (1-\lambda)f^2(y) + 2\lambda(1-\lambda)f(x)f(y)
$$
Now we manipulate this expression to obtain
$$
f^2(\lambda x+(1-\lambda)y) \leq  \lambda^2 f^2(x) + (1-\lambda)^2f^2(y) + 2\lambda(1-\lambda)f(x)f(y) + \left(-\lambda f^2(x) - (1-\lambda)f^2(y) + \lambda f^2(x) + (1-\lambda)f^2(y  \right)
$$
Combining the five first terms on the right hand side of this inequality yields
$$
f^2(\lambda x+(1-\lambda)y) \leq -\lambda(1-\lambda)\left(f(x)-f(y)\right)^2 + \lambda f^2(x) + (1-\lambda)f^2(y)
$$
and we immediately see that the first term on the right hand side is strictly negative as $0\leq\lambda\leq1$

This gives 
$$
f^2(\lambda x+(1-\lambda)y) \leq \lambda f^2(x) + (1-\lambda)f^2(y)

proving that the squared euclidean norm, is convex.

We know that the first orer necessary optimality condition is $\nabla f(x^*)=0$, so a minimizer $x^*$ must satisfy this. Further, we move forward under the assumption $x\notin \mathcal{M}$.
We compute the gradient of our objective function and equate to zero:

$$
We simply solve for $(x_1, x_2)$ and obtain

$$
\nabla f(x)=0 \iff \sum_{i\in \mathcal{M}} v_i\frac{x_j}{d_2(x, a^i)}=\sum_{i\in \mathcal{M}} v_i \frac{a_j^i}{d_2(x, a^i)} \hspace{2mm},\hspace{5mm}j=1,2
$$
which in yields the main iteration ofthe Weiszfeld scheme
$$
x_j=\frac{\sum_{i\in \mathcal{M}} v_i \frac{a_j^i}{d_2(x, a^i)}}{\sum_{i\in \mathcal{M}} v_i \frac{1}{d_2(x, a^i)}} \hspace{2mm},\hspace{5mm}j=1,2 \hspace{20mm}
$$
Now why is the iteration chosen like this? Firstly, it satisfies the first order necessary optimality condition. This combined with the convexity of the function, leads to a global minima $x^*$ of $f$ if and only if the gradient at $x^*$ is 0. 

We also note that this algorithm is very much like a gradient descent method. By manipulating the iteration scheme we see it clearly:
$$
x_j^{k+1}=\frac{\sum_{i\in \mathcal{M}} v_i \frac{a_j^i + x_j^k - x_j^k}{d_2(x, a^i)}}{\sum_{i\in \mathcal{M}} v_i \frac{1}{d_2(x, a^i)}} = \frac{\sum_{i\in \mathcal{M}} v_i \frac{x_j^k}{d_2(x, a^i)}}{\sum_{i\in \mathcal{M}} v_i \frac{1}{d_2(x, a^i)}} + \frac{\sum_{i\in \mathcal{M}} v_i \frac{(a_j^i - x_j^k)}{d_2(x, a^i)}}{\sum_{i\in \mathcal{M}} v_i \frac{1}{d_2(x, a^i)}}=x^k_j - \underbrace{\frac{1}{\sum_{i\in \mathcal{M}} v_i \frac{1}{d_2(x, a^i)}}}_{\alpha_k}\frac{\partial f}{\partial x_j}
\hspace{2mm},\hspace{5mm}j=1,2 \hspace{20mm}
$$
Since a gradient descent method on a differentiable and convex function converges to a global minima, the scheme converges to a global minima, and that is why the iteration is chosen in such a way.
